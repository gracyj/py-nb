```python
#NMF,Normalize
import pandas as pd
from sklearn.decomposition import NMF
import numpy as np
d1=pd.read_csv("./Musical artists/artists.csv",names=["Name"])
d2=pd.read_csv("./Musical artists/scrobbler-small-sample.csv")
print(d1.head())
print(d2.head())
```

```python
d1["Name"].unique()
```

```python
print(d1.shape)
print(d2.shape)
```

```python
d3 = d2.pivot_table(values="playcount",index=d2.artist_offset,columns="user_offset")
d3.fillna(0,inplace=True)
d3.head()
d3.set_index(d1["Name"], inplace=True)
print(d1.head())
print(d2.head())
```

```python
print(d3.head())
print(d3.shape)
```

```python
d3=d3.T
d3
```

```python
from sklearn.decomposition import NMF
model = NMF(n_components = 5)
model.fit(d3)
nmf_features = model.transform(d3)
print(nmf_features.shape)
```

```python
titles = list(d3.columns)
print(titles)
```

```python
from sklearn.preprocessing import normalize
norm_features = normalize(nmf_features)
df = pd.DataFrame(norm_features, index=titles)
datas = df.loc['Simon & Garfunkel']
similarities = df.dot(datas)
print(similarities.nlargest())
```
