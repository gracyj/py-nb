```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
cold = {0:'r',1:'g',2:'b'}
colr = [cold[i] for i in iris.target]
df.plot(kind="scatter",x="sepal length (cm)", y= "sepal width (cm)", c=colr)
```
